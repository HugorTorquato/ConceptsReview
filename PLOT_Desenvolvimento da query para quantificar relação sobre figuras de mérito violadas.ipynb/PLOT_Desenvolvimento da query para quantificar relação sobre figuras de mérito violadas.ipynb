```python
import pandas as pd
import matplotlib.pyplot as plt
```

Com base nos resultados descritos no "Desenvolvimento da query para quantificar relação sobre figuras de mérito violadas.pdf". Essa etapa consiste em inserir os dados no modelo de forma manual ( Poderia ser feito de outra forma mas são poucos... então não justifica ).

Nesse momento é criado um dataframe utilizando a lib pandas para ambos os casos, simulação com VW e simulação VV e VW.

Importante mencionar que este já é a soma de todas as simulações executadas.

image.png

```python
_15VW = {
    'Category': ['Overvoltage', 'Undervoltage', 'Overcurrent', 'Unbalance'],
    'Count': [157, 2, 0, 0]
}

_30VW = {
    'Category': ['Overvoltage', 'Undervoltage', 'Overcurrent', 'Unbalance'],
    'Count': [152, 0, 3, 0]
}

_50VW = {
    'Category': ['Overvoltage', 'Undervoltage', 'Overcurrent', 'Unbalance'],
    'Count': [113, 0, 1, 0]
}

_15VVVW = {
    'Category': ['Overvoltage', 'Undervoltage', 'Overcurrent', 'Unbalance'],
    'Count': [184, 44, 100, 0]
}

_30VVVW = {
    'Category': ['Overvoltage', 'Undervoltage', 'Overcurrent', 'Unbalance'],
    'Count': [173, 42, 89, 0]
}

_50VVVW = {
    'Category': ['Overvoltage', 'Undervoltage', 'Overcurrent', 'Unbalance'],
    'Count': [126, 37, 54, 0]
}

df_15VW   = pd.DataFrame(_15VW)
df_30VW   = pd.DataFrame(_30VW)
df_50VW   = pd.DataFrame(_50VW)

df_15VVVW = pd.DataFrame(_15VVVW)
df_30VVVW = pd.DataFrame(_30VVVW)
df_50VVVW = pd.DataFrame(_50VVVW)
```

#Plot

Vai seguir um padrão então vou separar cada um em blocos para ficar mais simples de identificar

15% + Controle VW

```python
df = pd.DataFrame(df_15VW)
df['Percentage'] = df['Count'] / df['Count'].sum() * 100

plt.figure(figsize=(8, 6))
plt.pie(df['Count'], startangle=140)
plt.title('15% + Controle VW')


legend_labels = [f'{cat} - {pct:.1f}%' for cat,
                 pct in zip(df['Category'], df['Percentage'])]

plt.legend(legend_labels, loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```

30% + Controle VW

```python
df = pd.DataFrame(df_30VW)
df['Percentage'] = df['Count'] / df['Count'].sum() * 100

plt.figure(figsize=(8, 6))
plt.pie(df['Count'], startangle=140)
plt.title('30% + Controle VW')


legend_labels = [f'{cat} - {pct:.1f}%' for cat,
                 pct in zip(df['Category'], df['Percentage'])]

plt.legend(legend_labels, loc="center left", bbox_to_anchor=(1, 0.5))

plt.show()
```
